```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('data.csv')
df
```

```python
reg= linear_model.LinearRegression()
reg.fit(df[['area','bathrooms','bedrooms','floors']],df.price)
```

```python
reg.predict([[2500,3,4,2]])
```

```python
slope = reg.coef_
slope
```

```python
intercept = reg.intercept_
intercept
```
